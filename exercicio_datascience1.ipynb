```
import pandas as pd
```

```
#Carregando um conjunto de dados pré-existente
df_kickstarter = pd.read_csv('/media/ks-projects-201801.csv')
df_kickstarter.head()
```

```
df_kickstarter.columns
```

```
categorias = df_kickstarter.groupby("main_category")
categorias.describe()
```

```
categorias['pledged'].sum().sort_values()
```

O que mais arrecadou foi a categoria Design.

## Realizar a separação entre os projetos que derão certo e os que não derão certo​

## Qual categoria mais superou o valor original estabelecido como meta?

```
a = categorias[['pledged','goal']].mean()
a['Comparação de valores'] = (a.pledged/a.goal) * 100
a.head(20)
```
